```python
import pandas as pd
import ast

# Load the dataset (replace with the actual path to your CSV file)
movies = pd.read_csv('tmdb_5000_movies.csv')

# Function to convert a stringified list of dictionaries into a comma-separated string of names
def convert_genres(genre_list):
    """Convert list of genre dictionaries into a comma-separated string of genre names."""
    if isinstance(genre_list, str):
        try:
            genre_list = ast.literal_eval(genre_list)  # Convert string to list
        except:
            return ''  # In case the string is malformed
    if isinstance(genre_list, list):
        return ', '.join([g['name'] for g in genre_list if 'name' in g])
    return ''

def convert_keywords(keyword_list):
    """Convert list of keyword dictionaries into a comma-separated string of keyword names."""
    if isinstance(keyword_list, str):
        try:
            keyword_list = ast.literal_eval(keyword_list)  # Convert string to list
        except:
            return ''  # In case the string is malformed
    if isinstance(keyword_list, list):
        return ', '.join([k['name'] for k in keyword_list if 'name' in k])
    return ''

# Apply the function to the 'genres' column to convert it into a comma-separated string of genre names
movies['genres'] = movies['genres'].apply(convert_genres)

# Apply the function to the 'keywords' column to convert it into a comma-separated string of keyword names
movies['keywords'] = movies['keywords'].apply(convert_keywords)

# Select the relevant columns: id, title, overview, genres, and keywords
result = movies[['id', 'title', 'overview', 'genres', 'keywords']]

# Display only the required columns (top 5 rows)
print(result.head())  # This will show only the required columns
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Ensure NaN values are handled in a single line
new_df['tags'] = new_df['tags'].fillna('')

# Initialize CountVectorizer with desired parameters
cv = CountVectorizer(max_features=5000, stop_words='english')

# Transform the 'tags' column into a numerical array
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        return " ".join(y)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[1])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movies):
    movies_index = new_df[new_df['title'] == movies].index[0]
    distances = similarity[movies_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Bang')
```
